```python
house_df['sqft_basement'].value_counts()
```

```python
X=house_df.drop('price',axis=1).values
y=house_df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
```

```python
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(X_train,y_train,epochs=400,validation_data=(X_test,y_test),batch_size=128)
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
X_train.shape
```

```python
loss_df.plot()
```

```python
model.evaluate(X_train,y_train)
```

```python
model.evaluate(X_test,y_test)
```

```python
predicted_values=model.predict(X_test)
```

```python
predicted_df=pd.DataFrame(predicted_values,columns=['predicted_price'])
predicted_df.head()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
np.sqrt(mean_squared_error(predicted_values,y_test))
```

```python
mean_absolute_error=mean_absolute_error(predicted_values,y_test)
```
